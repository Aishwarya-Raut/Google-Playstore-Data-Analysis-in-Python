#1. **Average Rating by Category**

# Objective:

This visualization will show the average rating of apps in different categories to help identify which categories tend to have higher or lower ratings.



```python
# Check if 'Category' and 'Rating' columns exist
if 'Category' in data.columns and 'Rating' in data.columns:
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns

    # Calculate average rating per category
    avg_rating = data.groupby('Category')['Rating'].mean().reset_index()

    # Sort categories by average rating for better visualization
    avg_rating = avg_rating.sort_values(by='Rating', ascending=False)

    # Set the figure size
    plt.figure(figsize=(12, 6))

    # Create a bar plot
    sns.barplot(x='Category', y='Rating', data=avg_rating, palette='coolwarm')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=90)

    # Add labels and title to the plot
    plt.xlabel('App Category', fontsize=12)
    plt.ylabel('Average Rating', fontsize=12)
    plt.title('Average Rating by Category', fontsize=16)

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Display the plot
    plt.show()
```

#**Insights**: We can observe which categories have the highest average ratings.

#**Implications**: Categories with higher ratings may indicate user satisfaction, providing opportunities for developers to focus on these categories.

#2. Distribution of Ratings

#Objective:
This visualization shows how ratings are distributed across all apps in the dataset. It helps us understand if the ratings are mostly positive or negative.

```python
#Visualization 1: Distribution of Ratings

plt.figure(figsize=(10, 6))
sns.histplot(data['Rating'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Ratings', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

#**Insights**: The histogram shows the spread of ratings. If the ratings are skewed, most apps will have ratings on the higher side or lower side.

#**Implications**: This helps developers understand user satisfaction trends and where they might need to improve.

#3. Relationship Between Installs and Rating

# Objective:
This scatter plot shows the relationship between the number of installs and the app rating. A log scale is applied to handle the large variation in the number of installs.


```python
# Clean the 'Installs' column
if 'Installs' in data.columns:
    # Ensure all values in the 'Installs' column are strings
    data['Installs'] = data['Installs'].astype(str)

    # Remove commas and plus signs, then convert to numeric
    data['Installs'] = data['Installs'].str.replace(',', '').str.replace('+', '').replace('Free', '0')
    data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')  # Convert to numeric, coercing errors to NaN

# Check if 'Installs' and 'Rating' columns exist
if 'Installs' in data.columns and 'Rating' in data.columns:
    # Drop rows with missing or invalid values in 'Installs' or 'Rating'
    data = data.dropna(subset=['Installs', 'Rating'])

    # Prepare the figure
    plt.figure(figsize=(12, 7))

    # Create a scatterplot
    sns.scatterplot(
        x='Installs',
        y='Rating',
        data=data,
        alpha=0.5,  # Transparency to avoid overcrowding
        color='blue',
        s=80  # Marker size
    )

    # Use log scale for the x-axis to handle large data range
    plt.xscale('log')

    # Add labels and title
    plt.title('Relationship Between Installs and Rating', fontsize=16)
    plt.xlabel('Number of Installs (Log Scale)', fontsize=14)
    plt.ylabel('Rating', fontsize=14)

    # Enhance grid visibility for better readability
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```

#**Insights**: This plot helps to understand if apps with more installs tend to have higher or lower ratings.

#**Implications**: Apps with more installs and higher ratings indicate strong user satisfaction, which could help guide marketing strategies.
